```python
# -*- coding: utf-8 -*-
"""
Created on Fri May 24 12:11:02 2024

@author: ahmed
"""
class PetriNet:
    def __init__(self):
        self.places = {
            'S0': 1,
            'S5': 0,
            'S10': 0,
            'S15': 0,
            'S20': 0
        }
        self.transitions = {
            'Deposit5_S0': {'pre': {'S0': 1}, 'post': {'S0': 0, 'S5': 1}},
            'Deposit5_S5': {'pre': {'S5': 1}, 'post': {'S5': 0, 'S10': 1}},
            'Deposit5_S10': {'pre': {'S10': 1}, 'post': {'S10': 0, 'S15': 1}},
            'Deposit5_S15': {'pre': {'S15': 1}, 'post': {'S15': 0, 'S20': 1}},
            'Deposit10_S0': {'pre': {'S0': 1}, 'post': {'S0': 0, 'S10': 1}},
            'Deposit10_S5': {'pre': {'S5': 1}, 'post': {'S5': 0, 'S15': 1}},
            'Deposit10_S10': {'pre': {'S10': 1}, 'post': {'S10': 0, 'S20': 1}},
            'Take15_S15': {'pre': {'S15': 1}, 'post': {'S15': 0, 'S0': 1}},
            'Take15_S20': {'pre': {'S20': 1}, 'post': {'S20': 0, 'S5': 1}},
            'Take20_S20': {'pre': {'S20': 1}, 'post': {'S20': 0, 'S0': 1}}
        }

    def fire_transition(self, choice):
        transition = None
        if choice == 'Deposit5' and self.places['S0'] == 1:
            transition = 'Deposit5_S0'
        elif choice == 'Deposit5' and self.places['S5'] == 1:
            transition = 'Deposit5_S5'
        elif choice == 'Deposit5' and self.places['S10'] == 1:
            transition = 'Deposit5_S10'
        elif choice == 'Deposit5' and self.places['S15'] == 1:
            transition = 'Deposit5_S15'
        elif choice == 'Deposit10' and self.places['S0'] == 1:
            transition = 'Deposit10_S0'
        elif choice == 'Deposit10' and self.places['S5'] == 1:
            transition = 'Deposit10_S5'
        elif choice == 'Deposit10' and self.places['S10'] == 1:
            transition = 'Deposit10_S10'
        elif choice == 'Take15' and self.places['S15'] == 1:
            transition = 'Take15_S15'
        elif choice == 'Take15' and self.places['S20'] == 1:
            transition = 'Take15_S20'
        elif choice == 'Take20' and self.places['S20'] == 1:
            transition = 'Take20_S20'

        if transition:
            if all(self.places.get(place, 0) >= count for place, count in self.transitions[transition]['pre'].items()):
                for place, count in self.transitions[transition]['pre'].items():
                    self.places[place] -= count
                for place, count in self.transitions[transition]['post'].items():
                    self.places[place] += count
                print(f'Transition {transition} fired.')
                return True
            else:
                print(f'Transition {transition} cannot be fired.')
                return False
        else:
            print("Invalid choice or insufficient tokens, please try again.")
            return False

    def __str__(self):
        return f'Current state: {self.places}'

def main():
    pn = PetriNet()
    while True:
        print(pn)
        print("Choose an action: ")
        print("1. Deposit 5")
        print("2. Deposit 10")
        print("3. Take 15 bar")
        print("4. Take 20 bar")
        print("5. Exit")
        choice = input("Enter your choice: ")
        if choice == '1':
            pn.fire_transition('Deposit5')
        elif choice == '2':
            pn.fire_transition('Deposit10')
        elif choice == '3':
            pn.fire_transition('Take15')
        elif choice == '4':
            pn.fire_transition('Take20')
        elif choice == '5':
            break
        else:
            print("Invalid choice, please try again.")

if __name__ == "__main__":
    main()
```
